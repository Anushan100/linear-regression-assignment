```python
import numpy as np
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import pickle
from pandas_profiling import ProfileReport
from sklearn import metrics
import matplotlib.pyplot as plt
import sklearn.datasets
```

```python
house_price=sklearn.datasets.load_boston()
```

```python
df=pd.DataFrame(house_price.data,columns=house_price.feature_names)
```

```python
df['price']=house_price.target
```

```python
x=df.drop(columns=['price'])
x
```

```python
y=df['price']
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=3)
```

```python
LINREG=LinearRegression()
```

```python
LINREG.fit(x_train,y_train)
```

```python
pred = LINREG.predict(x_test)
print(pred)
```

```python
print(f'MAE:{metrics.mean_absolute_error(y_test, pred)}')
print(f'MSE:{metrics.mean_squared_error(y_test, pred)}')
print(f'RMSE:{np.sqrt(metrics.mean_squared_error(y_test, pred))}')
```

```python

plt.scatter(y_test,pred)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Price vs Preicted Price")
plt.show()
```
